# Logistic Regression for Multiclass Classification: Theory

Logistic regression can be extended to handle multiclass classification problems, where the goal is to predict one of three or more possible classes. This is known as multinomial or multiclass logistic regression.

There are two main approaches:
- **One-vs-Rest (OvR):** The model trains a separate binary classifier for each class, treating that class as positive and all others as negative. The class with the highest probability is chosen as the prediction.
- **Multinomial (Softmax):** The model directly estimates the probability of each class using the softmax function, which generalizes the sigmoid function to multiple classes:

$$ P(y = k | x) = \frac{e^{z_k}}{\sum_{j=1}^K e^{z_j}} $$

where $z_k$ is the linear combination of features for class $k$ and $K$ is the total number of classes.

Multiclass logistic regression is widely used for tasks such as digit recognition, text classification, and disease diagnosis. It provides interpretable probabilities for each class and is efficient for both small and large datasets.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
digits=load_digits()
dir(digits)
```

```python
digits["data"][0]
```

```python
plt.gray()
plt.matshow(digits.images[11])
```

```python
X_Train,X_Test,Y_Train,Y_Test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_Test)
```

```python
len(X_Train)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_Train,Y_Train)
```

```python
model.score(X_Test,Y_Test)
```

```python
model.predict([digits.data[109]])
```
