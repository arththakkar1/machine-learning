# Single Linear Regression: Theory

Single linear regression is a statistical method used to model the relationship between two variables: a dependent variable (target) and an independent variable (predictor). The goal is to find the best-fitting straight line (regression line) that predicts the dependent variable based on the independent variable.

The equation for single linear regression is:

$$ y = mx + c $$

Where:
- $y$ is the predicted value (dependent variable)
- $x$ is the input value (independent variable)
- $m$ is the slope of the line
- $c$ is the intercept

The slope ($m$) represents how much $y$ changes for each unit increase in $x$. The intercept ($c$) is the value of $y$ when $x$ is zero.

Single linear regression is commonly used for prediction, trend analysis, and understanding relationships between variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("./Data/homeprice.csv")
df
```

```python
plt.xlabel("Area")
plt.ylabel("Prices")
plt.scatter(df["area"],df["price"],color="red",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
reg.predict([[100]])
```

```python
coe=reg.coef_
coe
```

```python
inter=reg.intercept_
inter
```
