# Train-Test Split: Theory

The train-test split is a fundamental technique in machine learning used to evaluate the performance of a model. It involves dividing a dataset into two separate sets:

- **Training set:** Used to train the model and learn the underlying patterns in the data.
- **Test set:** Used to evaluate the model's performance on unseen data and estimate how well it will generalize to new, real-world data.

A common split ratio is 70% for training and 30% for testing, but other ratios like 80/20 or 60/40 can also be used depending on the dataset size and problem.

The train-test split helps prevent overfitting, where a model performs well on training data but poorly on new data. By testing on unseen data, we get a more realistic measure of model accuracy and robustness.

In Python, the `train_test_split` function from `sklearn.model_selection` is commonly used to perform this split efficiently.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./Data/homeprice.csv")
df
```

```python
plt.scatter(df['area'],df['price'])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['area']],df['price'],test_size=0.2)
```

```python
modal=LinearRegression()
```

```python
modal.fit(X_train,Y_train)
```

```python
modal.predict([[4000]])
```

```python
modal.score(df[['area']],df['price'])
```
